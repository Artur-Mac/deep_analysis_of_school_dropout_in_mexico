```python
import pandas as pd
import numpy as np
import pingouin as pg
```

```python
path = "../data/raw/df03_Dataset_ENAPE.csv"
dfRaw = pd.read_csv(path)
dfRaw
```

```python
dfRaw.info()
```

```python
df = dfRaw[(dfRaw["EDAD"] >= 6) & (dfRaw["EDAD"] <= 18)]
df
```

```python
#mappings 9 to np.nan;

exceptions = ["nivel_edu", "EDAD", "nt_causa", "nins_causa", "q_hombres", "q_mujeres", "mejora_vida", "mejora_trabajo", "mejor_decisor"]

for i in df.columns:
    if i not in exceptions:
        df.loc[df[i] == 9, i] = 0
```

```python
compareColumn = set(np.array([0,1,2]))
print(compareColumn)

for i in df.columns:
    unique_values = set(df[i].values.astype(float))
    #unique_values = unique_values.astype(int)
    if unique_values.issubset(compareColumn):
        print(i)
        df.loc[df[i] == 0, i] = np.nan
        df.loc[df[i] == 2, i] = 0
    if 99 in unique_values or 98 in unique_values:
        df.loc[df[i] == 99, i] = np.nan
        df.loc[df[i] == 98, i] = np.nan

```

```python
# melhorando semantica das variaveis mejora_vida, mejora_trabajo, mejor_decisor
columnsToMap = ["mejora_vida", "mejora_trabajo", "mejor_decisor"]

mapping = {
    9: 0,
    4:-2,
    3:-1,
    2:1,
    1:2
}

for i in columnsToMap:
    if 9 in df[i].unique():
        df.loc[:, i] = df[i].map(mapping).astype("float")

```
